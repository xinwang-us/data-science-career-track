```python
import pandas as pd
```

```python
import sqlite3
from sqlite3 import Error

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e) 
    return conn
```

Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member formatted as a single column. 
Ensure no duplicate data, and order by the member name.

```python
pd.read_sql_query('select distinct f.name as facillity_name, m.firstname || " " || m.surname as member_name from members as m, facilities as f, bookings as b \
                 where f.name like "Tennis Court%" and f.facid = b.facid and b.memid = m.memid and b.memid <> 0 \
                 order by member_name', conn)
```

Q8: Produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30. 
    Remember that guests have different costs to members (the listed costs are per half-hour 'slot'), and
    the guest user's ID is always 0. Include in your output the name of the facility, the name of the member formatted 
    as a single column, and the cost. Order by descending cost, and do not use any subqueries. 

```python
pd.read_sql_query('select  f.name as facillity_name, m.firstname || " " || m.surname as member_name,  \
                           CASE WHEN b.memid=0 and b.slots*f.guestcost > 30 then b.slots*f.guestcost \
                                WHEN b.memid <> 0 and b.slots*f.membercost > 30 then b.slots*f.membercost \
                                END as cost \
                  from members as m, facilities as f, bookings as b \
                  where date(b.starttime) = "2012-09-14" and f.facid = b.facid and b.memid = m.memid and \
                  CASE WHEN b.memid=0 and b.slots*f.guestcost > 30 then b.slots*f.guestcost \
                       WHEN b.memid <> 0 and b.slots*f.membercost > 30 then b.slots*f.membercost END IS NOT NULL', conn)
```

Q9: This time, produce the same result as in Q8, but using a subquery.

```python
pd.read_sql_query('select *  \
                   from (select  f.name as facillity_name, m.firstname || " " || m.surname as member_name,  \
                                 CASE WHEN b.memid=0 and b.slots*f.guestcost > 30 then b.slots*f.guestcost \
                                      WHEN b.memid <> 0 and b.slots*f.membercost > 30 then b.slots*f.membercost END as cost \
                         from members as m, facilities as f, bookings as b \
                         where date(b.starttime) = "2012-09-14" and f.facid = b.facid and b.memid = m.memid ) as subq \
                   where subq.cost IS NOT NULL', conn)
```

Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, 
     sorted by revenue. Remember that there's a different cost for guests and members!

```python
pd.read_sql_query(' select facility_name, sum(cost) as total_revenue \
                    from (select f.name as facility_name, \
                                CASE WHEN b.memid = 0 THEN b.slots*f.guestcost \
                                     WHEN b.memid<>0 THEN b.slots*f.membercost END as cost \
                         from members as m, facilities as f, bookings as b \
                         where f.facid = b.facid and b.memid = m.memid ) as subq \
                    group by facility_name \
                    having sum(cost) < 1000 \
                    order by total_revenue', conn)
            
                  
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
pd.read_sql_query('select distinct m.firstname || " " || m.surname as member_name,  \
                          m1.firstname || " " || m1.surname as recomemndedby_name \
                   from members as m \
                        INNER JOIN members as m1 \
                        ON m.recommendedby = m1.memid \
                  where m.memid <> 0 \
                  order by m.surname, m.firstname', conn)
```

Q12: Find the facilities with their usage by member, but not guests 

```python
pd.read_sql_query('select f.name as facility_name, sum(b.slots) as usage \
                   from bookings as b, facilities as f \
                   where b.memid <> 0 and b.facid = f.facid \
                   group by facility_name', conn)
```

Q13: Find the facilities usage by month, but not guests 

```python
pd.read_sql_query('select strftime("%m", starttime) as month, f.name as facility_name, sum(b.slots) as usage \
                   from bookings as b, facilities as f \
                   where b.memid <> 0 and b.facid = f.facid \
                   group by month, facility_name', conn)
```
